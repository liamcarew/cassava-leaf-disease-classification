Now let's have a look at the number of images with poor focus by class in the training set:

```python
#function to return class label counts per class
def num_anomaly_by_class(cbb, cmd, cbsd, cgm, healthy):

  #create an instance of a dictionary in which the different image numbers will be stored
  anomaly_dict = {}

  anomaly_dict['CBB'] = [len(x) for x in cbb]
  anomaly_dict['CMD'] = [len(x) for x in cmd]
  anomaly_dict['CBSD'] = [len(x) for x in cbsd]
  anomaly_dict['CGM'] = [len(x) for x in cgm]
  anomaly_dict['Healthy'] = [len(x) for x in healthy]

  return anomaly_dict

#create a Pandas dataframe of the counts
poor_focus = [len(cbb_poor_focus), len(cmd_poor_focus), len(cbsd_poor_focus), len(cgm_poor_focus), len(healthy_poor_focus)]
classes = ['CBB', 'CMD', 'CBSD', 'CGM', 'Healthy']
poor_focus_df = pd.DataFrame(poor_focus, index=classes, columns = ['frequency'])

#plot the result
poor_focus_plt = poor_focus_df.plot(kind="bar", ylim=(0,120), legend=None, rot=0, color='#18cd85')
poor_focus_plt.set_xlabel('Class label', labelpad=10)
poor_focus_plt.set_ylabel('Number of images', labelpad=10)
```
